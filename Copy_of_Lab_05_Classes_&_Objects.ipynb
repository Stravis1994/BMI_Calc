### Your Answer

```python
# Student Name: Sean Travis
# Student Number: 1035888

from datetime import datetime

# TODO: Insert class definition:
class student:
  def __init__(self, name, age, course):
    self.name = name
    self.age = age
    self.course = course

# LIST
students= [
  student("Greg", 21, "Computer Science"),
  student("Lucy", 34, "Engineering"),
  student("Chris", 19, "Arts"),
  student("Sean", 31, "Computer Science")]

def print_student(*student):
  for i, student in enumerate(student, start=1):
    print(f"Name = {student.name}")
    print(f"Age = {student.age}")
    print(f"Course = {student.course}")
    print(f"Birth Year = {get_birth_year(student.age)}")
    print()

def get_birth_year(age):
  current_year = datetime.now().year
  birth_year = current_year - age
  return birth_year


for student in students:
  print_student(student)
```
